# Fractional 

# max profit 

```python
max_profit = df.sort_values(by=['profit'], ascending=False, ignore_index = True)
max_profit
```

```python
n = 7 
print(max_profit)
remain = max(df['profit'])
remainlist = []
cost = 0
for i in range (n):
    if remain - max_profit['w'][i] > 0:
        remain = remain - max_profit['w'][i]
        cost += max_profit["profit"][i]
        print(remain)
        remainlist.append(remain)
    elif remain >0 :
        cost += max_profit["profit"][i] * remain /  max_profit["w"][i]
        remain = 0
    else:    
        remainlist.append(0)
#ax_profit['XX'] = remainlist
print(max_profit)
print (f"cost equals to {cost}")
```
